```python
from pyspark.sql import SparkSession
from ast import alias
from pyspark.sql.functions import count
from pyspark.sql.functions import mean
from pyspark.sql.functions import max
from pyspark.sql.functions import desc
from pyspark.sql.functions import variance
from pyspark.sql.functions import round
from pyspark.sql.functions import min
from pyspark.sql.functions import year
from pyspark.sql.functions import month
from pyspark.sql.functions import avg
from pyspark.sql.window import Window
```

```python
spark = SparkSession.builder \
  .appName("EBAC LAB 29") \
  .getOrCreate()
```

```python
df_video = spark.read.parquet("/content/videos-preparados.snappy.parquet")
```

```python
df_video.groupBy("keyword").count().show()
```

```python
df_video.groupBy("keyword").agg(mean("Interaction").alias("media_interacao")).show()
```

```python
df_video.groupBy("keyword") \
  .agg(max("Interaction").alias("Rank Interaction")) \
  .orderBy(desc("Rank Interaction")) \
  .show()
```

```python
df_video.groupBy("Keyword").agg(
    mean("Views").alias("media_views"),
    variance("Views").alias("variancia_views")
).show()
```

```python
df_video.groupBy("Keyword").agg(
    round(mean("Views"), 0).alias("media_views"),
    round(min("Views"), 0).alias("min_views"),
    round(max("Views"), 0).alias("max_views")
).show()
```

```python
df_video.groupBy("keyword").agg(
    min("Published At").alias("primeiro_publicado"),
    max("Published At").alias("ultimo_publicado")
).show(truncate=False)
```

```python
total_titles_count = df_video.select("title").count()
unique_titles_count = df_video.select("title").distinct().count()

print("Total de títulos:", total_titles_count)
print("Total de títulos únicos:", unique_titles_count)
print("Diferença:", total_titles_count - unique_titles_count)
```

```python
df_video.withColumn("Ano", year("Published At")) \
  .groupBy("Ano") \
  .count() \
  .orderBy("Ano") \
  .show()
```

```python
df_video.withColumn("Ano", year("Published At")) \
    .withColumn("Mês", month("Published At")) \
    .groupBy("Ano", "Mês") \
    .count() \
    .orderBy("Ano", "Mês") \
    .show()
```

```python
df_likes_acumulativos = df_video.withColumn("Ano", year("Published At"))

janela = Window.partitionBy("Keyword").orderBy("Ano").rowsBetween(Window.unboundedPreceding, 0)

df_video.withColumn("Ano", year("Published At")) \
    .withColumn("Média_Ano", avg("Likes").over(Window.partitionBy("Keyword", "Ano"))) \
    .withColumn("Média_Acumulada_Likes", avg("Likes").over(janela)) \
    .select("Keyword", "Ano", "Média_Ano", "Média_Acumulada_Likes") \
    .distinct() \
    .orderBy("Keyword", "Ano") \
    .show()
```
